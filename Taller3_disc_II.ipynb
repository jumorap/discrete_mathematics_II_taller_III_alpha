# 1. 14a - 3) 
A vacuum pump removes one third of the remaining air in a cylinder with each stroke. Form an equation to represent this situation. After how many strokes is just $\frac{1}{1 000 000}$ of the initial air remaining?

Since each stroke removes $\frac{1}{3}$ of the air, then remaining $\frac{2}{3}$, and again is reduced by $\frac{1}{3}$, so, after n strokes we have left:<br>
$\Pi_{k = 0}^{n} \frac{2}{3} = \frac{2}{3} \cdot \frac{2}{3} \cdot \frac{2}{3} \cdot \frac{2}{3} \cdot \quad ... \quad \cdot \frac{2}{3}\quad$ $n$ times, until that $\Pi_{k = 0}^{n} \frac{2}{3} = \frac{1}{1000000}$

With this in mind, we can reduce the expression to an exponential expression, because the multiplying is not a dependent expression of $k$, as: $(\frac{2}{3})^{n}$, where $n$ tell us the number of times that shoud be stroked the vacuum to have $\frac{1}{1000000}$ of the total air. So:
<center>
    $(\frac{2}{3})^{n} = \frac{1}{1000000}$<br>
    $log_{\frac{2}{3}}((\frac{2}{3})^{n}) = log_{\frac{2}{3}}(\frac{1}{1000000})$<br>
    $n \cdot log_{\frac{2}{3}}(\frac{2}{3}) = log_{\frac{2}{3}}(\frac{1}{1000000})$<br>
    $n \cdot 1 = log_{\frac{2}{3}}(\frac{1}{1000000})$<br>
    $n = log_{\frac{2}{3}}(10^{-6})$<br>
    $n = -6 log_{\frac{2}{3}}(10) \approx 34$ times
</center>
